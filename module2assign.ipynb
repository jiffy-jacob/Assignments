```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel('/content/drive/MyDrive/ABC Company.xlsx')
df
```

```python
print(df.Height)
```

```python
#Preprocessing step
# Replace Height column with random integers between 150 and 180
df['Height'] = np.random.randint(150, 181, size=len(df))
print(df['Height'])
```

```python
#Distribution of employees across each team
employees_team=df["Team"].value_counts()
print(employees_team)
```

```python
#percentage split of teams relative to the total number of employees
percent_split=(employees_team/len(df))*100
print(percent_split)
```

```python
#Visualization of employee distribution across teams
plt.figure(figsize=(8,6))
sns.countplot(data=df,x='Team')
plt.xticks(rotation=90)
plt.title("Employee Distribution by Team")
plt.xlabel("Team")
plt.ylabel("Count")
```

#Inference
1. It is evident that highest number of employees belong to New Orleans Pelicans Team.
2. Second highest number of employees are in Memphis Grizlies team.
3. Lowest number of employees are in Orlando Magic team as well as in Minnesota Timberwolves team.

```python
#Segregate employees based on their positions within the company.
employees_position=df["Position"].value_counts()
print(employees_position)
```

```python
#Visualisation
plt.figure(figsize=(8,6))
sns.countplot(data=df,x='Position')
plt.xticks(rotation=90)
plt.title("Employee Distribution by Position")
plt.xlabel("Position")
plt.ylabel("Count")
```

#Inferences
1. Highest number of employees belong to SG position
2. Second highest number of employees in PF position
3. Lowest number of employees in C position.

```python
#Identify the predominant age group among employees.
employees_age=df['Age'].value_counts()
print(employees_age)
```

```python
bins=[22,26,30,34,38,42,46,50]
labels=['22-26', '26-30', '30-34', '34-38', '38-42', '42-46', '46-50']
df['Age_Group']=pd.cut(df['Age'],bins=bins,labels=labels)
print(df['Age_Group'])
```

```python
#Visualization
plt.figure(figsize=(8,6))
sns.countplot(data=df,x='Age_Group')
plt.xticks(rotation=90)
plt.title("Employee Distribution by age group")
plt.xlabel("Age Group")
plt.ylabel("Count")
```

#Inference
1. Highest number of employees are in the age group of 22 to 26
2. Ssecond highest number of employees are in the age group of 26 to 30.
3. Lowest number of age groups are in the age group of 38 to 42.

```python
# To find which team have the highest salary expenditure.
team_salary=df.groupby('Team')['Salary'].sum()
print(team_salary)
print("Highest salary is for ",team_salary.idxmax(),"team")
```

```python
#Visualisation
plt.figure(figsize=(8,6))
sns.barplot(x=team_salary.index,y=team_salary.values)
plt.xticks(rotation=90)
plt.title("Salary Expenditure by Team")
plt.xlabel("Team")
plt.ylabel("Salary")
```

#Inferences
1. Highest salary is for  Cleveland Cavaliers team.
2. Lowest salary is for Philadelphia 76ers team.

```python
# To find which position have the highest salary expenditure
salary_position=df.groupby('Position')['Salary'].sum()
print(salary_position)
print("Highest salary is for ",salary_position.idxmax(),"position")
```

```python
#Visualisation
plt.figure(figsize=(8,6))
sns.barplot(x=salary_position.index,y=salary_position.values)
plt.xticks(rotation=90)
plt.title("Salary Expenditure by Position")
plt.xlabel("Position")
plt.ylabel("Salary")
```

#Inferences
1. Highest salary is for  C position
2. Lowest salary is for SG position

```python
# finding if there's any correlation between age and salary
correlation1=df['Age'].corr(df['Salary'])
print("correlation coefficient is",correlation1)
```

```python
#Visualization
plt.figure(figsize=(8,6))
sns.scatterplot(data=df,x='Age',y='Salary')
plt.title("Correlation between Age and Salary")
plt.xlabel("Age")
plt.ylabel("Salary")
```

#Inferences
1. The correlation coefficient we got is 0.21400941226570971, which means it has a weak positive linear relation between age and salary of employees
2. This means as age increases, salary tends to increase slightly, but there's no strong, consistent pattern.
